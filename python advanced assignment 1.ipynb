# Q2. Where does an inheritance search look for an attribute?

The whole point of a namespace tool like the class statement is to support name inheritance. 

In Python, inheritance happens when an object is qualified, 

and involves searching an attribute definition tree (one or more namespaces).

Every time you use an expression of the form object.attr where object is an instance or class object,

Python searches the namespace tree at and above object, for the first attr it can find. Because lower definitions in the tree override higher ones, 
inheritance forms the basis of specialization

Instance attributes are generated by assignments to self attributes in methods.

Class attributes are created by statements (assignments) in class statements.

Superclass links are made by listing classes in parentheses in a class statement header
